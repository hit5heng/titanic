```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
from keras import layers
```

```python
tf.__version__
```

```python
data1 = pd.read_csv("./data1.csv")
```

```python
data1
```

```python
x_train = data1[['Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin','p1', 'p2', 'p3', 'e1', 'e2', 'e3']]

y_train = data1['Survived'].values.reshape(len(data1),1)

print(x_train.shape,y_train.shape)
```

```python
data2 = pd.read_csv("./data_test.csv")
x_test = data2[['Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin','p1', 'p2', 'p3', 'e1', 'e2', 'e3']]

data3 = pd.read_csv('./data/gender_submission.csv')
y_test = data3['Survived'].values.reshape(len(data3),1)
print(x_test.shape,y_test.shape)
```

```python
model = keras.Sequential()
```

```python
model.add(layers.Dense(1,input_dim=12,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['acc']
)
```

```python
history = model.fit(
    x_train, y_train,
    epochs = 300,
    validation_data = (x_test,y_test)
)
```

```python
history.history.keys()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(300), history.history.get("loss"))
```

```python
plt.plot(range(300), history.history.get("acc"))
```
